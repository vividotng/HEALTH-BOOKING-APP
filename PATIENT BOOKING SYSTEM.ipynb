```python
import csv
class Patient:
    def __init__(self,patient_id, name, age, gender, phone,
                 pain_level, report_time, first_visit,
                 complaint, address, emergency_contact):
        self.patient_id = patient_id
        self.name = name
        self.age = age
        self.gender = gender
        self.phone = phone
        self.pain_level = pain_level
        self.report_time = report_time
        self.first_visit = first_visit
        self.complaint = complaint
        self.address = address
        self.emergency_contact = emergency_contact
        
    #list to store all the patients
patients =[]

while True:

    print("\nEnter Patient Details")

    patient_id = input("Patient ID: ")
    name = input("Name: ")
    age = int(input("Age: "))
    gender = input("Gender: ")
    phone = input("Phone Number: ")
    pain_level = input("Pain Level: ")
    report_time = input("Report Time: ")
    first_visit = input("First Visit (Yes/No): ")
    complaint = input("Complaint: ")
    address = input("Address: ")
    emergency_contact = input("Emergency Contact: ")

    patient = Patient(
        patient_id, name, age, gender, phone,
        pain_level, report_time, first_visit,
        complaint, address, emergency_contact
    )

    patients.append(patient)

    another = input("\nAdd another patient? (yes/no): ").lower()

    if another == "no":
        break
#fo the number of patients
# Number of patients to register
filename = "patients.csv"

with open(filename, "w", newline="") as file:
    writer = csv.writer(file)

    # Header row
    writer.writerow([
        "Patient ID",
        "Name",
        "Age",
        "Gender",
        "Phone",
        "Pain Level",
        "Report Time",
        "First Visit",
        "Complaint",
        "Address",
        "Emergency Contact"
    ])

    # Write patient data
    for patient in patients:
        writer.writerow([
            patient.patient_id,
            patient.name,
            patient.age,
            patient.gender,
            patient.phone,
            patient.pain_level,
            patient.report_time,
            patient.first_visit,
            patient.complaint,
            patient.address,
            patient.emergency_contact
        ])

print(f"\n{len(patients)} patient(s) successfully saved to '{filename}'.")
```
